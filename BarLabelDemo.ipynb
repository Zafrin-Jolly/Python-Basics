```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#defining the data
N=5
menMeans=(20,35,20,35,-27)
womenMeans=(25,32,34,20,-25)
menStd=(2,3,4,1,2)
womenStd=(3,5,2,3,3)
ind=np.arange(N) #the x location
width=0.5 #the width of the bars
```

```python
fig,ax = plt.subplots()

p1 = ax.bar(ind, menMeans, width, yerr=menStd, label='Men',color='red')
p2=ax.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd, label='Women',color='blue')

ax.axhline(0, color='black',linewidth=1)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1','G2','G3','G4','G5'))
ax.legend() #label with label type center instead of the default 'edge'
#ax.bar_label(p1,label_type='center')
#ax.bar_label(p2,label_type='center')
#ax.bar_label(p2)
plt.show()
```

```python
#horizontal bar chart
# fixing random state for reproducibility
np.random.seed(19680000)
#Example data 
people=('Jolly','Tisha','Shoumik','Sakib','Dipro','Tazul')
y_pos= np.arange(len(people))
performance = 3 +5 * np.random.rand(len(people))
error=np.random.rand(len(people))
fig,ax=plt.subplots()
hbars=ax.barh(y_pos, performance,xerr=error,align='center',color='red')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis() #labes to read top_to_bottom
ax.set_xlabel('Hour')
ax.set_title('How many hour did you study')# Label with specially formatted floats
#ax.bar_label(hbars, fmt='%.2f')
ax.set_xlim(right=10)  # adjust xlim to fit labels

plt.show()
```
